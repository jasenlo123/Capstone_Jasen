```python
urls = []
dates = []
for page in tree.all_pages():
    urls.append(page.url)
    dates.append(page.last_modified)
```

```python
#transpose
zippedList =  list(zip(urls,dates))

#format into dataframe
df = pd.DataFrame(zippedList, columns = ['urls' ,'date'])

#drop nan for date published
df = df.dropna()

#set time stamps
year_2019 = datetime.datetime(2019, 1, 1, 1, 1, 1, tzinfo=tzoffset(None, 28800))
year_2020 = datetime.datetime(2020, 1, 1, 1, 1, 1, tzinfo=tzoffset(None, 28800))


#drop articles before 1/1/2019
indexNames = df[(df['date'] <= year_2019)].index
df.drop(indexNames, inplace=True)

#drop articles after 1/1/2020
indexNames = df[(df['date'] >= year_2020)].index
df.drop(indexNames, inplace=True)

#pickle dataset
with open('scmp_df.pkl', 'wb') as f:
    pickle.dump(df, f)
```

```python
#TESTING TIME OF THE TITLE SCRAP

urls = df['urls'].values
#sampling = random.choices(urls, k=10)

headlines = []

start = time.time()

for url in urls:
    r1 = requests.get(url)
    cont = r1.content
    soup1 = BeautifulSoup(cont, 'lxml')
    
    #find headline in html beautiful soup
    [title] = soup1.find_all('title')
    
    #remove tags
    title = title.get_text()
    
    #Remove 'South China Morning Post' in string
   # title = title.split("|")[0]
    
    #Remove 'The Straits Times' in string
    title = title.split(" -")[0]
    
    headlines.append(title)

end = time.time()
print(end - start)
```

```python
headlines
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA

sia = SIA()

gt_sentiment = []
for text in headlines:
    print(sia.polarity_scores(text)["compound"])
    gt_sentiment.append(sia.polarity_scores(text)["compound"])

np.mean(gt_sentiment)
```
